# Princípio Fundamental da Contagem (PFC)

##  O que é?

O Princípio Fundamental da Contagem diz:

> Se um processo ocorre em etapas independentes, multiplicamos as possibilidades de cada etapa.

---

## Intuição

Se você tem escolhas em sequência, você **multiplica**.

---



##  Exemplo 1 — Roupas

- 3 camisetas  
- 2 calças  

Quantos looks diferentes?

$$
3 \times 2 = 6
$$

---



##  Exemplo 2 — Moeda + Dado

- Moeda → 2 opções (cara, coroa)  
- Dado → 6 opções  

Total de possibilidades:

$$
2 \times 6 = 12
$$

---



##  Conexão com Probabilidade

A probabilidade é dada por:

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

👉 O PFC nos ajuda a encontrar o **total de casos possíveis**.

---



## 📌 Exemplo aplicado

Qual a probabilidade de tirar **cara e depois o número 6**?

- Total de possibilidades: 12  
- Casos favoráveis: 1  

$$
P = \frac{1}{12}
$$

## 1. Conceito fundamental

Probabilidade mede a chance de um evento acontecer:

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

Exemplo:
Dado → pares = {2,4,6}

P(par) = 3/6 = 0.5


```python

# exemplo simples

pares = [2,4,6]
dado = [1,2,3,4,5,6]

probabilidade = len(pares) / len(dado)
print(f' A probabilidade de sair um número par é: {probabilidade*100} %')
```


## 5. Eventos dependentes (intuição)

Se retirar carta sem reposição → muda a probabilidade

Isso é MUITO importante para IA (dados mudam!)


EXEMPLO 1: CARTAS

“Se eu tiro uma carta e NÃO devolvo… o baralho muda.”

```python
cartas = ["❤️", "❤️", "♠", "♠"]
```

Qual a probabilidade de tirar duas cartas ❤️ seguidas sem reposição?

1ª carta:

* 2/4 = 1/2

2ª carta:

agora só sobram 3 cartas
apenas 1 é ❤️

* 1/3

P = 1/2 * 1/3 = 1/6

```python
import random

cartas = ["❤️", "❤️", "♠", "♠"]

sucesso = 0
n = 1000

for _ in range(n):
    baralho = cartas.copy()
    c1 = random.choice(baralho)
    baralho.remove(c1)

    c2 = random.choice(baralho)

    if c1 == "❤️" and c2 == "❤️":
        sucesso += 1

sucesso / n
```

## 7. Probabilidade em IA

IA retorna algo assim:

Gato → 0.8  
Cachorro → 0.2  

Ela NÃO tem certeza → ela tem probabilidade


```python

#Código
import numpy as np
prob = np.array([0.8,0.2])

classes = ["gato", "cacchorro"]

classes[np.argmax(prob)]
```


## 8. Interpretação MUITO importante

0.8 NÃO significa certeza  
Significa: “alta chance baseada nos dados”


## 11. Desafio IA

Crie um sistema:

Entrada: probabilidades  
Saída: decisão da IA


```python

# espaço aluno


def decisao_ia(probabilidade):
    if probabilidade >= 0.9:
        return "A IA está muito confiante: decisão positiva."
    elif probabilidade >= 0.7:
        return "A IA considera provável: decisão favorável."
    elif probabilidade >= 0.5:
        return "A IA está incerta: decisão neutra ou cautelosa."
    else:
        return "A IA considera improvável: decisão negativa."

# Exemplos de uso
print(decisao_ia(0.95))  # Alta confiança
print(decisao_ia(0.75))  # Provável
print(decisao_ia(0.55))  # Incerto
print(decisao_ia(0.30))  # Improvável
```

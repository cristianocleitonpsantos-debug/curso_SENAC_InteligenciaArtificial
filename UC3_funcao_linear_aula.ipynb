## 1. Revisão rápida: o que é uma função?

Uma função é uma regra que associa cada valor de entrada a um único valor de saída.

Exemplo:

- entrada: quantidade de quilômetros rodados
- saída: preço da corrida

Se o preço depende da distância, podemos representar isso por uma função.

---

## 2. Função linear

A forma geral da função linear que vamos usar nesta aula é:

\[
f(x) = ax + b
\]

Onde:

- `x` = valor de entrada;
- `f(x)` = valor de saída;
- `a` = taxa de variação / inclinação da reta;
- `b` = valor inicial / ponto onde a reta corta o eixo y.

### Interpretação
- Se `a > 0`, a função cresce.
- Se `a < 0`, a função decresce.
- Se `b = 0`, a reta passa pela origem.

## 7. Aplicação em IA: previsão com função linear

Agora vem a conexão com Inteligência Artificial.

Imagine que queremos prever o preço de uma casa com base no tamanho dela.

Temos os seguintes dados:

| Tamanho (m²) | Preço (mil R$) |
|-------------|----------------|
| 50          | 100            |
| 60          | 120            |
| 70          | 140            |

Perceba que existe um padrão:

- a cada 10 m², o preço aumenta 20 mil;
- isso equivale a 2 mil por m².

Podemos escrever:

\[
preço = 2x
\]

ou, no formato geral:

\[
f(x) = 2x + 0
\]

### O que isso tem a ver com IA?
Nesse caso:

- `x` = variável de entrada (**feature**)
- `f(x)` = valor previsto
- `a` = peso (**weight**)
- `b` = viés (**bias**)

> Em Machine Learning, um dos objetivos é encontrar automaticamente os melhores valores de `a` e `b`.

```python
# Simulando uma previsão simples com função linear

def preco_casa(tamanho):
    return 2*tamanho  # em mil reais

for tamanho in [45, 55, 80, 100]:
    print(f"Tamanho: {tamanho} m² -> Preço previsto: R$ {preco_casa(tamanho)*1000:,.2f}")
```

## 8. Regressão linear: uma IA aprendendo a reta

Agora vamos usar uma biblioteca para ajustar automaticamente uma reta aos dados.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Dados
X = np.array([50, 60, 70]).reshape(-1, 1)   # entrada
y = np.array([100, 120, 140])               # saída

# Modelo
modelo = LinearRegression()
modelo.fit(X, y)

print("Coeficiente a:", modelo.coef_[0])
print("Intercepto b:", modelo.intercept_)

# Previsão
novo_valor = np.array([[85]])
previsao = modelo.predict(novo_valor)

print(f"Preço previsto para 85 m²: {previsao[0]:.2f} mil reais")
```

### Interpretação
O modelo encontrou automaticamente uma função do tipo:

\[
f(x) = ax + b
\]

Ou seja, ele aprendeu a melhor reta para representar os dados.

### Onde vocês usarão isso no futuro?
Ao construir IA, vocês usarão esse raciocínio em:

- regressão linear;
- classificação com soma ponderada;
- redes neurais (camadas fazem operações lineares);
- análise de dados e previsão;
- ajuste de parâmetros em modelos.

> Mesmo quando o modelo é mais complexo, a lógica matemática de entrada, peso e saída continua aparecendo.

## 7. Tratamento de texto e padronização

### Exercício 13
Padronize a coluna `bolsista` para ficar apenas com:
- `Sim`

### Exercício 14

### Exercício 15

Exemplo:
- `brasília` → `Brasília`


```python
# Resolva aqui
df['bolsista'] = (
    df['bolsista']
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        'sim': 'Sim',
        'não': 'Não',
        'nao': 'Não'
    }))
```

```python
df['uf'] = df['uf'].astype(str).str.upper()
```

```python
df['cidade'] = df['cidade'].astype(str).str.title()
```

## 9. Ordenação, contagem e agrupamento

### Exercício 19
Ordene os alunos pela coluna `media`, do maior para o menor.

### Exercício 20
Conte quantos alunos existem em cada curso.

### Exercício 21
Mostre a média de `nota1` por curso.

### Exercício 22
Mostre a média de faltas por turma.


```python
# Resolva aqui
import pandas as pd

df = pd.read_excel('base_exercicios_alunos_pandas.xlsx')

# garantir que nota1/nota2 sejam numéricas
df['nota1'] = pd.to_numeric(df['nota1'], errors='coerce')
df['nota2'] = pd.to_numeric(df['nota2'], errors='coerce')

# criar media
df['media'] = (df['nota1'] + df['nota2']) / 2

# ordenar do maior para o menor
df_ordenado = df.sort_values(by='media', ascending=False)
df_ordenado
```

```python
df['curso'].value_counts(dropna=False)
```

```python
df['nota1'] = pd.to_numeric(df['nota1'], errors='coerce')
df.groupby('curso', dropna=False)['nota1'].mean()
```

```python
df['faltas'] = pd.to_numeric(df['faltas'], errors='coerce')
df.groupby('turma', dropna=False)['faltas'].mean()
```
